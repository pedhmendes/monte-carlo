##O Modelo

###Urnas $A$ e $B$


Considere duas urnas, $A$ e $B$, e um certo número $N$ de esferas, numeradas de 1 até $N$. Inicialmente todas esferas estão em $A$. Escolhemos então uma esfera aleatória e a trocamos de urna. Queremos saber a probabilidade $P_l(n)$ de ter $n$ esferas na urna $A$ em um tempo $l$.

Em um certo tempo $l$, existem $n$ esferas em $A$. A probabilidade do número de esferas diminuir para $n-1$ é a mesma de escolher uma esfera de $A$ e mover para $B$, isso  é, $n/N$. De forma similar, a probabilidade do número de esferas em A aumentar para $n+1$ é a probabilidade de escolher uma esfera de $B$ e mover para $A$, isso é, $(N-n)/N$. Escrevendo em forma de probababilidade de transição $T(n,m)$, de $m$ para $n$ obtemos.

\begin{aligned}
    T(n-1, n) = \frac{n}{N}
\end{aligned}

\begin{aligned}
    T(n+1, n) = \frac{N-n}{N}
\end{aligned}

É importante notar que para $N$ esferas sempre teremos $N+1$ estados possíveis. Isso acontece porque o estado 0, onde todas esferas estão na urna $B$, também é possível e deve ser considerado. Abaixo podemos ver um diagrama para $N = 6$ esferas e suas probabilidades de transição. O diagrama foi retirado do livro *Stochastic Dynamics and Irreversibility*, Tânia Tomé e Mário J. de Oliveira.


Screenshot from 2021-03-01 18-09-35.png

Note que para $N = 6$ esferas possuímos sete estados, pois o estado onde não tem nenhuma esfera em $A$ existe e deve ser considerado.

Porém o caso estudado aqui será um caso mais geral. Nesse caso nós introduzimos uma probabilidade $p$ de que o número de esferas em $A$ não mude, ou seja, de permanecer no mesmo estado. Escrevendo as probabilidades de transição obtemos


\begin{aligned}
    T(n-1, n) = q \frac{n}{N}
\end{aligned}

\begin{aligned}
    T(n, n) = p
\end{aligned}

\begin{aligned}
    T(n+1, n) = q\frac{N-n}{N}
\end{aligned}

onde $q = 1 - p$ é a probabilidade de mudar de estado. Note que o modelo Ehrenfest original é recuperado fazendo $p = 0$. É interessante notar que a matriz tem um formato tridiagonal, onde as diagonais seguem as três ultimas equações apresentadas. Veja:

\begin{aligned}
    T = 
\begin{pmatrix}
p &q\frac{1}{N}&0& \ldots & \cdots& \ldots &  \ldots & 0\\
q \frac{N-0}{N}&p&q\frac{2}{N}& \ldots &\cdots & \ldots & \ldots& 0\\
0 &q \frac{N-1}{N}&p& \ldots & \cdots & \ldots& \ldots & 0\\
\vdots & \vdots & \vdots & \ddots & q\frac{n}{N} &  \ldots & \ldots & \vdots\\
\vdots & \vdots & \vdots & \ldots & p & \ldots& \ldots & \vdots \\
\vdots & \vdots & \vdots & \ldots & q\frac{N-n}{N} & \ddots & \ldots & \vdots\\
\vdots & \vdots & \vdots & \vdots & \vdots & \cdots  & \ddots& \vdots\\
0&0&0& \ldots & \cdots& \ldots & \ldots &  p\\
\end{pmatrix}
\end{aligned}

onde $n$ é o estado. Outra informação importante de ser salientada é que a soma das probabilidades de cada coluna deve ser igual a 1. Isso significa que a soma das probabilidades de não mudar a configuração, de remover e de adicionar uma esfera na urna deve ser a probabilidade total, 1.

Inserindo $T(n, m)$ na evolução temporal

\begin{aligned}
    P_{l+1}(n) = \sum_{n=0}^N T(n, m) P_l(m)
\end{aligned}

obtemos

\begin{aligned}
    P_{l+1}(n) = q \left(1 - \frac{n-1}{N} \right)P_l(n-1) +
                p P_l(n) + q \frac{n+1}{N} P_l(n+1)
\end{aligned}

que é uma equação valida para $n = 0, 1, 2, \cdots, N$, se fizermos $P_l(N+1) = 0$ e $P_l(-1) = 0$. A equação é resolvida para a condição inicial $P_0(n) = \delta_{n,N}$, que significa todas as esferas começam na urna $A$. 

Vemos que a condição de probabilidade estacionária cumpre a equação

\begin{aligned}
    P(n) = \left(1 - \frac{n-1}{N} \right)P_l(n-1) + \frac{n+1}{N} P_l(n+1)
\end{aligned}

cuja solução é

\begin{aligned}
    P(n) = 2^{-N} \left( \begin{array}{c}N\\n\end{array} \right)
\end{aligned}

Agora vamos calcular o número médio de esferas na urna $A$. Sabendo que o primeiro momento é dado por

\begin{aligned}
    X_l = \sum_{n=0}^N n P_l(n)
\end{aligned}

obtemos

\begin{aligned}
    X_l = \frac{N}{2} + \frac{N}{2}\left( 1- \frac{2}{N} \right)^l
\end{aligned}

Podemos ver que $X_l$ tende a $N/2$. Similarmente para o segundo momento

\begin{aligned}
    Y_l = \sum_{n=0}^N n^2 P_l(n)
\end{aligned}

cujo resultado é

\begin{aligned}
    Y_l = \frac{N(N+1)}{4} + \frac{N^2}{2}\left(1 - \frac{2}{N}q \right)^l +
          \frac{N(N-1)}{4}\left(1 - \frac{4}{N}q \right)^l
\end{aligned}

Para definirmos a variância utilizamos o primeiro e o segundo momento da forma

\begin{aligned}
    Y_l - X_l^2 = \frac{N}{4} - \frac{N^2}{4}\left(1 - \frac{2}{N}q \right)^l + \frac{N(N-1)}{4}\left(1 - \frac{4}{N}q \right)^l
\end{aligned}

Podemos ver que a variância tende a $N/4$.


###Demonstrações

Agora iremos demonstrar três equações citadas acimas: ($i$) que a solução é o binômio citado, ($ii$) o primeiro e o ($iii$) segundo momento.

($i$) começando pela solução, aplicamos o binômio no sistema estacionário.

\begin{aligned}
    2^{-N} \left( \begin{array}{c}N\\n\end{array} \right) =
      \left(1 - \frac{n-1}{N} \right)2^{-N} \left( \begin{array}{c}N\\n-1\end{array} \right) + 
      \frac{n+1}{N} 2^{-N} \left( \begin{array}{c}N\\n+1\end{array} \right)
\end{aligned}

Cortando os $2^{-N}$

\begin{aligned}
    \left( \begin{array}{c}N\\n\end{array} \right) =
      \left(1 - \frac{n-1}{N} \right) \left( \begin{array}{c}N\\n-1\end{array} \right) + 
      \frac{n+1}{N} \left( \begin{array}{c}N\\n+1\end{array} \right)
\end{aligned}

Abrindo o binômio

\begin{aligned}
    \left( \begin{array}{c}N\\n\end{array} \right) =
     \left(\frac{N-n+1}{N} \right) \frac{N!}{(n-1)!(N-n+1)!} 
     + 
     \frac{n+1}{N}\frac{N!}{(n-1)!(N-n-1)!} 
\end{aligned}

Rearranjando os fatoriais

\begin{aligned}
    \left( \begin{array}{c}N\\n\end{array} \right) =
     \frac{n}{N}\frac{N!}{n!(N-n)!} 
     + 
     \frac{N-n}{N}\frac{N!}{n!(N-n)!}    
\end{aligned}
Cortando os binômios
\begin{aligned}
    1 = \frac{n}{N} + \frac{N-n}{N} 
\end{aligned}

\begin{aligned}
    1 = 1
\end{aligned}

Assim $2^{-N} \left( \begin{array}{c}N\\n\end{array} \right)$ é solução e é a única.

($ii$) seguindo para o cálculo da média fazemos

\begin{aligned}
    X_{l+1} = \sum^N_{n=0} n P_{l+1}
\end{aligned}

\begin{aligned}
    X_{l+1} = \sum^N_{n=0} n \left[ q\left(1- \frac{n-1}{N}\right)P_l(n-1) + p P_l(n) + q \frac{n+1}{N} P_l(n+1) \right]
\end{aligned}

\begin{aligned}
    X_{l+1} = \sum^N_{n=0} n \left[ qP_l(n-1) - q \left(\frac{n-1}{N} \right)P_{l}(n-1)+ p P_l(n) + q \frac{n+1}{N} P_l(n+1) \right]
\end{aligned}

reescrevendo o primeiro termo:

\begin{aligned}
    q \sum^N_{n=0} n P_l(n-1) =  
    q \sum^{N-1}_{k=0} (k+1) P_l(k)
\end{aligned}

\begin{aligned}
    q \sum^{N-1}_{k=0} (k+1) P_l(k) = 
    q \left[
    \sum^{N}_{k=0} k P_l(k) + \sum^{N-1}_{k=0} P(k) -NP_l(k)
    \right]
\end{aligned}


\begin{aligned}
    q \sum^{N-1}_{k=0} (k+1) P_l(k) = 
    q \left[
    X_l + \sum^{N-1}_{k=0} P(k) -NP_l(k)
    \right]
\end{aligned}
Reescrevendo o terceiro termo

\begin{aligned}
    p \sum^N_{n=0} n P_l(n) =  p X_{l}
\end{aligned}

Reescrevendo o segundo e o quarto termo

\begin{aligned}
    \frac{q}{N} \sum^{N-1}_{n=0} n \left[ (n+1)P_l(n+1) - (n-1)P_l(n-1) \right]  &=\\
    &= \frac{q}{N} \left[ -2P_l(1) - 4P_l(2) - \cdots - 2(N-1)P_l(N-1)\right]
    +
    q \left[ (N-1) P_l(N) + (N+1) P(N+1) \right]
    \\ 
    &=\frac{q}{N}(-2) \sum^{N-1}_{n=0} n P_l(n) + q(N-1)P_l(N) \\
    &=-\frac{2q}{N}X_l + qP_l(N) + qNP_l(N)
\end{aligned}

Retornando aos termos obtidos

\begin{aligned}
    X_{l+1} = (q+p)X_l + q\left[\sum^{N-1}_{k=0}P_l(k) + P_l(N) \right] - \frac{2q}{N}X_l
\end{aligned}
Lembrando que o somatório da probabilidade de todos os estados deve ser 1, chegamos em
\begin{aligned}
    X_{l+1} = X_l + q \left( 1 - \frac{2}{N}X_l \right)
\end{aligned}

Agora supondo que o estado inicial seja o que todas esferas estejam na urna $A$, isto é

\begin{aligned}
    P_0(N) = 1, &&  P_0(n) = 0 && \forall n \neq N
\end{aligned}

A média em $l=0$ será

\begin{aligned}
    X_0 = \sum^N_{n=0} n P_0(n) = nP_0(N) = N
\end{aligned}

Para os tempos seguidos obteremos

\begin{aligned}
    X_1 &= X_0 + \left( 1 - \frac{2}{N}X_0 \right) &= N-q \\
    X_2 &= X_1 + \left( 1 - \frac{2q}{N} \right) + q &= \frac{N}{2}\left( 1 - \frac{2q}{N} \right)^2 + \frac{N}{2} \\
    X_3 &= X_2 + \left( 1 - \frac{2q}{N} \right) + q &= \frac{N}{2}\left( 1 - \frac{2q}{N} \right)^3 + \frac{N}{2} \\
\end{aligned}

Assim vemos que

\begin{aligned}
    X_l & = \frac{N}{2}\left( 1 - \frac{2q}{N} \right)^l + \frac{N}{2} \\
\end{aligned}

que é exatamente a equação exibida anteriormente.


($iii$) para o cálculo da variância precimos provar a equação do segundo momento. Para isso fazemos

\begin{aligned}
    Y_{l+1} = \sum^N_{n=0} n^2 P_{l+1}
\end{aligned}

\begin{aligned}
    Y_{l+1} = \sum^N_{n=0} n^2 \left[ q\frac{N- n+1}{N} P_l(n-1) +
                p P_l(n) + q \frac{n+1}{N} P_l(n+1) \right]
\end{aligned}

fazendo passos similares aos feitos para o primeiro momento:

\begin{aligned}
    \sum^N_{n=0} n^2 p P_l(n) &= q \left[ Y_l + \sum_{k=0}^{N-1}(P_l(k) + 2k P_l(k)) - N^2 P_l(N) \right] \\
    &= q \left[ Y_l + \sum_{k=0}^{N-1}2k P_l(k)) - N^2 P_l(N) \right]
\end{aligned}

\begin{aligned}
    \sum^N_{n=0} n^2 p P_l(n) = p Y_l
\end{aligned}

\begin{aligned}
    \frac{q}{N} \sum^N_{n=0} n^2 \left[(n+1)P_l(n+1) - (n-1)P_l(n-1) \right] &=  \frac{q}{N}\left[ -4P_l(1) - 16P_l(2) - \cdots - 4(n-1)^2P_l(n-1) \right] + \frac{q}{N} \left[ (n-1)P_l(N) + N^2(N+1)P_l(N+1) \right] \\
    &=- \frac{4q}{N}Y_l + 4qNP_l(N) + q(N-1)^2P_l(N)
\end{aligned}

Retornando a equação do segundo momento e agrupando os termos

\begin{aligned}
    Y_{l+1} &= qY_l + q \sum_{k=0}^{N-1} (2k+1)P_l(k) -qN^2P_l(N) + pY_l - \frac{4q}{N} Y_l + 4qNP_l(N) + q(N-1)^2P_l(N) \\
    &= \left(1 - \frac{4q}{N}\right)Y_l + q\sum_{n=0}^{N}P_l(n) + 2q\left[ \sum_{k=0}^{N-1}kP_l(k +NP_l(N) \right]
\end{aligned}

Note que o somatório do segundo termo é 1 e os termos dentro dos colchetes equivale a $X_l$, logo

\begin{aligned}
    Y_{l+1} = \left(1 - \frac{4q}{N}\right)Y_l + q(1 + 2X_l)
\end{aligned}

Substituindo o resultado de $X_l$

\begin{aligned}
    Y_{l+1} = \left(1 - \frac{4q}{N}\right)Y_l + q\left(1 + N + \left(1-\frac{2q}{N}\right)^l\right)
\end{aligned}

Novamente fazendo

\begin{aligned}
    P_0(N) = 1 ,&& P_0(n) = 0 ,&& \forall n \neq N
\end{aligned}

Obtemos que o segundo momento é

\begin{aligned}
    Y_0 = N^2
\end{aligned}

assim

\begin{aligned}
    Y_1 &= \left(1-\frac{4q}{N}\right)^2 + q(2N+1) = N^2 -2Nq +q \\
    Y_2 &= \left(1-\frac{4q}{N}\right)^2N^2 + q(2N+1)\left(1- \frac{4q}{N}\right) + q\left(1+N+N\left(1-\frac{2q}{N}\right)\right) \\
\end{aligned}

A para um $l$ qualquer obtemos

\begin{aligned}
    Y_l &= \frac{N}{4}(N+1) + \frac{N^2}{2}\left(1-\frac{2}{N}q\right)^l + \frac{N}{4}(N-1)\left(1 - \frac{4q}{N}\right)^l
\end{aligned}

Com o o primeiro e o segundo momento podemos obter a expressão da variância da forma

\begin{aligned}
    Y_l - X_l^2 &= \frac{N}{4}(N+1) + \frac{N^2}{2}\left(1-\frac{2}{N}q\right)^l + \frac{N}{4}(N-1)\left(1 - \frac{4q}{N}\right)^l - 
    \left( \frac{N}{2}\left( 1 - \frac{2q}{N} \right)^l + \frac{N}{2}  \right) \\
    &=\frac{N}{4} + \left(1 - \frac{4q}{N} \right)^l \frac{N}{4}(N-1) - \left(1 - \frac{2q}{N} \right)^l \frac{N^2}{4}
\end{aligned}

Por fim obtemos

\begin{aligned}
    Y_l - X_l^2 &= \frac{N}{4} +\frac{N(N-1)}{4} \left(1 - \frac{4q}{N} \right)^l  - \frac{N^2}{4}\left(1 - \frac{2q}{N} \right)^l 
\end{aligned}

Que é a equação apresentada anteriormente

###Expansão em Autovetores

Chamaremos {$\lambda_i$} de autovalores e {$\psi_i$} e {$\phi_i$} de autovetores à direita e à esquerda, respectivamente. Note que {$\psi_i$} são vetores coluna e {$\phi_i$} são vetores linha. Os autovetores formam um conjunto completo ortonormalizado, isto é, respeitam duas condições: 


\begin{aligned}
    \sum_i \psi_i \phi_i = I
\end{aligned}

onde $I$ é a matriz identidade. E


\begin{aligned}
    \phi_i \psi_j = \delta_{i,j}
\end{aligned}

onde $\delta_{i,j}$ é a função delta de Kronecker. 
 Podemos escrever $T^{l}$ como

\begin{aligned}
    T^l = T^l I = T^l\sum_i \psi_i \phi_i = \sum_i T^l\psi_i \phi_i
\end{aligned}
E como por definição:

\begin{aligned}
    T \psi_{i} = \lambda_{i} \psi_{i}
\end{aligned}

Temos que

\begin{aligned}
    T^l = \sum_i \lambda^l\psi_i \phi_i
\end{aligned}

A partir de $T^l$ podemos calcular a probabilidade de estar no estado $n$ no tempo $l$ da seguinte forma

\begin{aligned}
    P_l = T^l P_0 = \sum_i \lambda^l\psi_i \phi_i P_0
\end{aligned}

onde $P_0$ é o estado inicial.


A probabilidade estacionária $P$ é o autovetor de autovalor um (1), sendo assim podemos reescrever $P_l$ como

\begin{aligned}
    P_l = P + \sum_{i \neq 1} \lambda^l_i \psi_i \phi_i P_0
\end{aligned}

onde $\lambda_1^l = 1$ e $\psi_1 = P$



##Simulação

###Forma Matricial

Como vamos lidar com matrizes foi escolhido utilizar Python 3.x. Começamos importando as bibliotecas necessárias.

```
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.linalg as linalg
from random import randint

%matplotlib inline
```

Definindo $N$ igual ao número total de esferas. Aqui é tomado um valor pequeno, para que os cálculos não demorem e também para que as matrizes não fiquem desconfiguradas na tela. Primeiro é montada uma matriz nula de $N+1$ estados 

```
N = 3
L = N+1

T = np.zeros((L,L))
print(T)

T.shape
```

Seguindo então as equações demonstradas podemos aplicar as probabilidades correspondentes aos estados, montar a matriz de probabilidade de transição e verificar que a soma das probabilidades é igual a 1

```
p = 0.5
q = 1 - p

for j in range(L):
    for i in range(L):
        if i == j:
            T[i,j] = p
        if i == j-1:
            T[i, j] = q*(j/N)
        if i == j + 1:
            T[i, j] = q*(N-j)/N

print(T)
```

Vamos então calcular os autovalores, os autovetores à direita e à esquerda

```
vals , rvec = linalg.eig(T, left=False, right=True)
lvec = linalg.inv(rvec)

vals = vals.real

print("Autovalores:\n%s\n"%vals)
print("Autovetores à direita: (colunas)\n%s\n"%rvec)
print("Autovetores à esquerda: (linhas)\n%s"%lvec)
```


Definindo aqui algumas funções que iremos usar depois. As funções serão utilizadas para calcular o produto interno e o produto externo

```
#PRODUTO INTERNO DOS AUTOVETORES <phi|chi>
def prod_int(j,k):
	return np.dot(lvec[j,:], rvec[:,k])

#PRODUTO EXTERNO DOS AUTOVETORES |psi><chi|
def prod_ext(j,k):
	return np.dot(rvec[:,j][np.newaxis].transpose(),lvec[k,:][np.newaxis])
```

Podemos ver que nosso sistema respeita a primeira condição
\begin{aligned}
    \sum_i |\psi_i><\phi_i| = I
\end{aligned}

```
I = np.zeros((L,L))

for i in range(L):
    I += prod_ext(i,i)

print(I)
```

Respeita a segunda condição também
\begin{aligned}
   <\chi_i|\psi_k> = \delta_i,k
\end{aligned}

```
kronecker = np.zeros((L,L))
for j in range(L):
	for k in range(L):
		kronecker[j][k] = prod_int(j,k)
  
print(kronecker)
```

Calculando

\begin{aligned}
     T^l = \sum_i \lambda_i^l \psi_i \phi_i
\end{aligned}

para todos os tempo de l=0 a l=49.

```
Tl = np.zeros((L,L,50)) # matriz Tl para vários tempos, matriz 3D
for l in range(50): 
  for i in range(L):
  	Tl[:,:,l] = Tl[:,:,l] + (vals.real[i]**l)*prod_ext(i,i)
l=5 # Escolha o tempo de quiser de 0 a 49
print("Exemplo de Tl para um l específico: ") #Nesse caso escolhemos 5
print(Tl[:,:,l]) #Matriz de probabilidade no tempo l = 5
```

Calculando a probalidade dos estados para um tempo l
\begin{aligned}
    P_l = T^l P_0
\end{aligned}

```
P0 = np.zeros(L)
P0[L-1] = 1.0 # O estado inicial é o que todas as bolinhas estão em A

Pl = np.dot(Tl[:,:,l], P0)

print(Pl)
```

Converge para o estado estacionário?

\begin{aligned}
     P = |\psi_1><\phi_1|P_0 = \psi_1 \phi_1 P_0
\end{aligned}
onde \psi_1 e \chi_1 se referem aos autovetores cujo autovalor é 1

```
P = np.dot(prod_ext(0,0),P0)
print("Probabilidade estacionária:\n%s"%P)
print("Pl com l=49:\n%s"%np.dot(Tl[:,:,49],P0))
```

###Sistema

Simulação do sistema, deve coincidir com os resultados analiticos para $N =3 $. 

```
N = 3
passos = 40000 #100000
p = 0.5 #prob de ficar
q = 1 - p #prob de troca
A = N # Quantas bolas há em A 
B = 0 # Quantas bolas há em B
histn = np.zeros((N+1))
media = np.zeros((passos))
desvio = np.zeros((passos))

def trocar_bolas(A,B):
	n = randint(0, N-1)	# escolhe uma bola
	A = A + trocas[int(bolas[n]),1]
	B = B + trocas[int(bolas[n]),0]
	bolas[n] = bolas[n] + troca_bola[int(bolas[n])]
	return A,B

def media_n(l):
	media = 0
	for i in range(N+1):
		media += i*histn[i];
	return media/l

def media_quadrada(l):
	media2 = 0
	for i in range(N+1):
		media2 += (i**2)*histn[i];
	return media2/l

trocas = np.array([[-1,+1],[+1,-1]])
troca_bola = np.array([+1,-1])

bolas = np.ones(N) # Todas as bolas em A

for i in range(1,passos,1):
	r = (randint(0, 1000))/1000 # gerador
	if(r <= q):
		A,B = trocar_bolas(A,B)
	histn[A] += 1;
	media[i] = media_n(i)
	desvio[i] = media_quadrada(i) - (media_n(i)**2)
```

Gráfico da média em função dos passos. Como é possível ver converge para $N/2 = 3/2$

```
fig, ax = plt.subplots(figsize=(16, 9))
plt.plot(np.arange(0,passos,1),media,label="Média")#,color="k")
plt.hlines(N/2.0, 0,passos, color = "red")#resultado analítico

plt.xlabel("Passos")
plt.ylabel("Média")
plt.legend(loc='best')

plt.show()
```
